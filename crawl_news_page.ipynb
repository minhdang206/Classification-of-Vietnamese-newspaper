## Import các thư viện cần thiết

```python
import requests
import re
import os
from urllib.parse import urlsplit
from bs4 import BeautifulSoup
from IPython.display import clear_output

from google.colab import drive
drive.mount('/content/drive')
```

```python
def url_to_file_name(url):
    url = str(url).replace('/', '_')
    return re.findall(r'vnexpress.net.(\w.+)',url)[0] + ".html"

def download(file_store,url):
    print("Saving html file,...")
    content=requests.get(url).text    
    filename=os.path.join(file_store, url_to_file_name(url))    

    with open(filename, 'w',encoding='utf-8') as f:
        f.write(content)
```

```python
dir_folder = r'/content/drive/MyDrive/[FINAL]_KPW/data_html'
url_main_page = 'https://vnexpress.net'

dict_folder_label = {
    'khoa_hoc'    : ['Khoa Học','https://vnexpress.net/khoa-hoc'],
    'doi_song'    : ['Đời Sống','https://vnexpress.net/doi-song'],
    'the_thao'    : ['Thể Thao','https://vnexpress.net/the-thao'],
    'du_lich'     : ['Du Lịch','https://vnexpress.net/du-lich'],
    'suc_khoe'    : ['Sức Khỏe','https://vnexpress.net/suc-khoe'],
    'kinh_doanh'  : ['Kinh Doanh','https://vnexpress.net/kinh-doanh'],
    'phap_luat'   : ['Pháp Luật','https://vnexpress.net/phap-luat'],
    'giai_tri'    : ['Giải Trí','https://vnexpress.net/giai-tri'],
    'giao_duc'    : ['Giáo Dục','https://vnexpress.net/giao-duc'],
}

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}
```

```python
for label in dict_folder_label.items():
  check_done_label = False
  url = label[1][1]
  while check_done_label == False:
    folder_store = os.path.join(dir_folder, label[1][0])

    soup_html_main_page = BeautifulSoup(requests.get(url, headers = headers).text , 'lxml')
    list_link_news = soup_html_main_page.select('[class="title-news"] > a , [class="title_news"] > a')
    for idx, link_new in enumerate(list_link_news):
      # clear_output()
      download(folder_store,link_new.attrs["href"])
      print(f"Label - {label[0]} - Crawled {idx+1} of {len(list_link_news)} in page {url}")
    print(f"Label - {label[0]} - Saved data of {url}")
    
    next_page = soup_html_main_page.select_one('[class="btn-page next-page"]')
    if next_page:
      url = url_main_page + next_page.attrs["href"]
    else:
      check_done_label = True
      print(f"Done label {label[0]}")
```
